```python
# Define the available routes and their characteristics
routes = {
    "W": {"distance_km": 300, "travel_time": 25, "traffic": "Medium"},
    "X": {"distance_km": 200, "travel_time": 15, "traffic": "Low"},
    "Y": {"distance_km": 150, "travel_time": 20, "traffic": "Medium"},
    "Z": {"distance_km": 250, "travel_time": 10, "traffic": "High"}
}

# Map traffic levels to numeric scores
traffic_scores = {
    "Low": 3,
    "Medium": 2,
    "High": 1
}

# Compute a utility score for each route
def compute_score(route):
    dist = route["distance_km"]
    time = route["travel_time"]
    traffic = traffic_scores[route["traffic"]]
    
    # Weighted negative utility: lower is worse
    score = -(dist*2 + time*3.5 + traffic*1)
    return score

# Store computed scores
route_scores = {}

for route_name, route_info in routes.items():
    route_scores[route_name] = compute_score(route_info)

# Select the route with the highest utility
optimal_route = max(route_scores, key=route_scores.get)

# Display scores for all routes
for route, score in route_scores.items():
    print(f"Route {route}: {score}")

print(f"\nOptimal Route: {optimal_route} with utility {route_scores[optimal_route]}")
```

```python
#TASK 3
#environment state
environment = {
    "Section 1": "Healthy",
    "Section 2": "Wilted"
}

#agent state (unknown at start)
internal_state = {
    "Section 1": "Unknown",
    "Section 2": "Unknown"
}

#agent position
agent_location = "Section 1"

#Agent observes the current section
def observe(section):
    return environment[section]

#Watering plants
def water(section):
    environment[section] = "Healthy"

#change section
def move(section):
    if section == "Section 1":
        return "Section 2"
   
    if section == "Section 2":
        return "Section 1"

# numbre of runs
time_steps = 6

for t in range(time_steps):
    print(f"\n--- Time Step {t+1} ---")
    print(f"Agent location: {agent_location}")

    #initiate
    perception = observe(agent_location)
    internal_state[agent_location] = perception

    #think for action
    if perception == "Wilted":
        action = "Water"
        water(agent_location)
    else:
        action = "Move"
        agent_location = move(agent_location)

    print(f"Action taken: {action}")
    print("Agent internal state:", internal_state)
    print("Actual environment:", environment)
```
